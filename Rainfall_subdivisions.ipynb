```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'/content/rainfall in india 1901-2015.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#plotting the mean annual rainfall and note down your observations regarding the same
a=df.groupby("YEAR").mean()['ANNUAL'].plot(ylim=(800,2200), color='b', marker='o', linestyle='-', linewidth=2, figsize=(12,4))
```

```python
#Finding the moving average taking into consideration a period of 10 years and plotting
df['MA10'] = df.groupby('YEAR').mean()['ANNUAL'].rolling(10).mean()
```

```python
df.MA10.plot(color='b',linewidth=2)
```

```python
#Aesthetic of the plot
plt.xlabel('year',fontsize=20)
plt.ylabel('Annual Rainfall (in mm)', fontsize=20)
plt.title('Annual Rainfall in India from Year 1901 to 2015',fontsize=25)
plt.xlim(1900,2020)
a.tick_params(labelsize=15)
plt.grid()
plt.ioff()
plt.show()
```

```python
# Plotting a bar plot of the mean seasonal rainfall
df[['SUBDIVISION','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].groupby("SUBDIVISION").mean().sort_values('Jun-Sep').plot.bar(width=0.5, edgecolor='k',align='center',stacked=True,figsize=(16,8))
```

```python
# Aesthetic of the plot
plt.xlabel('subdivision',fontsize=30)
plt.ylabel('Rainfall(in mm)', fontsize=20)
plt.title('Rainfall in Subdivisions of India',fontsize=25)
a.tick_params(labelsize=15)
plt.grid()
plt.ioff()
plt.show()
```

```python
# Plotting a bar plot of the mean seasonal rainfall
df[['SUBDIVISION','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].groupby("SUBDIVISION").mean().sort_values('Jun-Sep').plot.bar(width=0.5, edgecolor='k',align='center',stacked=True,figsize=(16,8))
```

```python
#Analyze the rainfall through years in kerala and note down your observations
#plotting the average annual rainfall
kerala= df[df.SUBDIVISION == 'KERALA']
```

```python
ax=kerala.groupby("YEAR").mean()['ANNUAL'].plot(ylim=(1000,5000), color='b', marker='o',linewidth=2,linestyle='-',figsize=(12,8))
```

```python
#setting up the aesthetics of the plot
plt.xlabel('Year',fontsize=20)
plt.ylabel('kerala Annual Rainfall (in mm)',fontsize=20)
plt.title('kerala Annual rainfall from year 1901 to 2015',fontsize=25)
ax.tick_params(labelsize=15)
plt.grid()
plt.ioff()
plt.show()
```
